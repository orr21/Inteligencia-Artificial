```python
import matplotlib.pyplot as plt

# coordenadas de los puntos que queremos dibujar
x = [50, 50, 0, 0]
y = [50, 0, 50, 0]
labels = ['T: 4', 'T: 3', 'T: 2', 'T: 1']

x2 = [20,15,40]
y2 = [35,10,20]
labels2 = ['S: WB', 'S: WD', 'S: NY']

# crear el gráfico
fig, ax = plt.subplots()

# dibujar los puntos con etiquetas
ax.scatter(x, y, c='blue')
for i, txt in enumerate(labels):
    ax.annotate(txt, (x[i] - 1, y[i] + 0.8))

ax.scatter(x2, y2, c='black')
for i, txt in enumerate(labels2):
    ax.annotate(txt, (x2[i] - 1, y2[i] + 0.8))

# draw a circle around the origin (0,0) with radius 16
for i, txt in enumerate(labels2):
    circle = plt.Circle((x2[i], y2[i]), 2, color='red', fill=False)
    ax.add_artist(circle)
    circle = plt.Circle((x2[i], y2[i]), 6, color='yellow', fill=False)
    ax.add_artist(circle)
    circle = plt.Circle((x2[i], y2[i]), 12, color='orange', fill=False)
    ax.add_artist(circle)

# agregar etiquetas de los ejes y título
ax.set_xlabel('long')
ax.set_ylabel('lat')
ax.set_title('Mapa de coordenadas')

# mostrar el gráfico
plt.show()
```
